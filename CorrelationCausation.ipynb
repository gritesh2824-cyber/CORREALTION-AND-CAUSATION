#CORRELATION
Correlation is a statistical relationship that tells us **how strongly two variables move together.**
   
 It gives answers of questions like this :
1) When X increases, does Y also increase?
2) When X decreases, does Y decrease?
3) Are X and Y unrelated?

  # Types of Correlation
  1. Positive Correlation (+) :

  Both variables move in the same direction.

Example:

Height â†‘ â†’ Weight â†‘

Study time â†‘ â†’ Marks â†‘

**Meaning:** If one increases, the other increases too.

 2. Negative Correlation (â€“) :

   Both variables move in the opposite direction.

Example:

Speed of fan â†‘ â†’ Room temperature â†“

Price â†‘ â†’ Demand â†“

**Meaning:** If one increases, the other decreases.

3. Zero Correlation (0) :

   There is no relationship between the variables.

Example:

Height and IQ

Shoe size and marks

**Meaning:** One variable changes, but the other stays random.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate sample data
np.random.seed(42)
X = np.random.rand(100) * 10  # Independent variable
Y = 2 * X + np.random.randn(100) * 2  # Dependent variable with some noise, showing positive correlation

df = pd.DataFrame({'Variable_X': X, 'Variable_Y': Y})

display(df.head())
```

Now, let's visualize the relationship between `Variable_X` and `Variable_Y` using a scatter plot. A scatter plot is excellent for showing the distribution of two variables and can visually suggest correlation.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Variable_X', y='Variable_Y', data=df)
plt.title('Scatter Plot of Variable_X vs. Variable_Y (Positive Correlation Example)')
plt.xlabel('Variable_X')
plt.ylabel('Variable_Y')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

To quantify the correlation, we can calculate the Pearson correlation coefficient, which ranges from -1 (perfect negative correlation) to +1 (perfect positive correlation), with 0 indicating no linear correlation.

```python
correlation_coefficient = df['Variable_X'].corr(df['Variable_Y'])
print(f"Pearson Correlation Coefficient between Variable_X and Variable_Y: {correlation_coefficient:.2f}")

plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Variables')
plt.show()
```

##CAUSATION :
Causation means one event directly produces another event.
A cause is something that makes an effect happen.

**In short,**

If change in variable A produces a change in variable B,
 then A causes B.

 example : If you heat water â†’ it boils.

  Here heating is the cause, boiling is the effect.

  # Types of Causation :

1)Direct Causation :

   A directly causes B.

**Example:** Pushing a switch â†’ turns ON the light.

2)Indirect Causation :

   A causes C through B.

**Example:**

  [Exams â†’ Stress â†’ Lack of sleep]

Exams indirectly cause lack of sleep.

3)Multiple Causation :

 Many factors together cause one outcome.

**Example:** Accident caused by

*   Speed
*   Bad road
* Mobile use

4)Bidirectional Causation

A affects B and B affects A.

**Example:**

Stress â†‘ â†’ Productivity â†“

Productivity â†“ â†’ Stress â†‘

5)Probabilistic Causation

Cause increases the probability, not guarantee.

**Example:** Smoking increases risk of cancer.

```python
np.random.seed(42) # Set random seed for reproducibility

# Generate 100 data points for Study_Hours between 0 and 10
study_hours = np.random.rand(100) * 10

# Generate Exam_Scores based on Study_Hours with a linear relationship and noise
# Let's assume a base score of 30, and each hour of study adds 5 points, plus some noise
exam_scores = 30 + (5 * study_hours) + (np.random.randn(100) * 10) # Added more noise for realism

# Ensure scores don't go below 0 or above 100
exam_scores = np.clip(exam_scores, 0, 100)

# Create a pandas DataFrame
df_causal = pd.DataFrame({
    'Study_Hours': study_hours,
    'Exam_Scores': exam_scores
})

# Display the first few rows of the created DataFrame
display(df_causal.head())
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Study_Hours', y='Exam_Scores', data=df_causal, label='Data Points', color='blue')
sns.regplot(x='Study_Hours', y='Exam_Scores', data=df_causal, scatter=False, color='red', line_kws={'linestyle': '--', 'linewidth': 2}, label='Regression Line')
plt.title('Causal Relationship: Study Hours vs. Exam Scores')
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
